```python
# function that extracts the category of the venue
def get_category_type(row):
    try:
        categories_list = row['categories']
    except:
        categories_list = row['venue.categories']
        
    if len(categories_list) == 0:
        return None
    else:
        return categories_list[0]['name']
```

```python
def division(n, d):
    return n / d if d else 0

def getNearbyVenues(names, latitudes, longitudes, radius=400):
    
    neighborhoods_venues=[]
    for name, lat, lng in zip(names, latitudes, longitudes):
        print(name)
            
        # create the API request URL
        url = 'https://api.foursquare.com/v2/venues/search?client_id={}&client_secret={}&ll={},{}&v={}&categoryId={}&radius={}&limit={}'.format(CLIENT_ID, CLIENT_SECRET, lat, lng, VERSION, categoryId, radius, LIMIT)
        url2 = 'https://api.foursquare.com/v2/venues/search?client_id={}&client_secret={}&ll={},{}&v={}&categoryId={}&radius={}&limit={}'.format(CLIENT_ID, CLIENT_SECRET, lat, lng, VERSION, categoryId2, radius, LIMIT)
        url3 = 'https://api.foursquare.com/v2/venues/search?client_id={}&client_secret={}&ll={},{}&v={}&categoryId={}&radius={}&limit={}'.format(CLIENT_ID, CLIENT_SECRET, lat, lng, VERSION, categoryId3, radius, LIMIT)
        url4 = 'https://api.foursquare.com/v2/venues/search?client_id={}&client_secret={}&ll={},{}&v={}&categoryId={}&radius={}&limit={}'.format(CLIENT_ID, CLIENT_SECRET, lat, lng, VERSION, categoryId4, radius, LIMIT)
        url6 = 'https://api.foursquare.com/v2/venues/search?client_id={}&client_secret={}&ll={},{}&v={}&categoryId={}&radius={}&limit={}'.format(CLIENT_ID, CLIENT_SECRET, lat, lng, VERSION, categoryId6, radius, LIMIT)
        url7 = 'https://api.foursquare.com/v2/venues/search?client_id={}&client_secret={}&ll={},{}&v={}&categoryId={}&radius={}&limit={}'.format(CLIENT_ID, CLIENT_SECRET, lat, lng, VERSION, categoryId7, radius, LIMIT)
        
        # make the GET request
        results = requests.get(url).json()
        venues = results['response']['venues']
        
        results2 = requests.get(url2).json()
        venues2 = results2['response']['venues']
        
        results3 = requests.get(url3).json()
        venues3 = results3['response']['venues']
        
        results4 = requests.get(url4).json()
        venues4 = results4['response']['venues']
        
        results6 = requests.get(url6).json()
        venues6 = results6['response']['venues']
        
        results7 = requests.get(url7).json()
        venues7 = results7['response']['venues']
        
        #ratio = division(len(venues2) * 2 + len(venues4), len(venues) * 5 + len(venues3) * 2)
        #ratio = 1 + len(venues) * 5 + len(venues3) * 2 / 1 + len(venues2) * 2 + len(venues4)

        
        # return only relevant information
        neighborhoods_venues.append([
            name, 
            lat, 
            lng, 
            len(venues),
            len(venues2),
            len(venues3),
            len(venues4),
            len(venues6),
            len(venues7)
            ])

    nearby_venues = pd.DataFrame(neighborhoods_venues)
    nearby_venues.columns = ['Neighborhood', 
                  'Neighborhood Latitude', 
                  'Neighborhood Longitude', 
                  'Num pools',
                  'Num gyms',
                  'Num ice cream shops',
                  'Num supermarkets',
                  'Num coworkings',
                  'Num monuments'
                ]
    
    return(nearby_venues)
```

```python
madrid_venues = getNearbyVenues(names=neighborhoods['name'],
                                   latitudes=neighborhoods['lat'],
                                   longitudes=neighborhoods['lng']
                                  )
```

```python
print(madrid_venues.shape)
ratio = [((1 + row['Num pools'] * 10 + row['Num ice cream shops'] * 5 + row['Num monuments']) * 3 / (1 + (row['Num gyms'] * 2 + row['Num supermarkets'] + row['Num coworkings']) * row['Num supermarkets'])) for index, row in madrid_venues.iterrows()]

madrid_venues['Ratio'] = ratio

madrid_neighb = madrid_venues.sort_values('Ratio')
madrid_neighb
```

```python
# set number of clusters
kclusters = 4

madrid_grouped_clustering = madrid_neighb[['Num pools', 'Num gyms', 'Num ice cream shops', 'Num supermarkets', 'Num coworkings', 'Num monuments']]

# run k-means clustering
kmeans = KMeans(n_clusters=kclusters, random_state=0).fit(madrid_grouped_clustering)

# check cluster labels generated for each row in the dataframe
kmeans.labels_[0:10] 
```

```python
madrid_grouped_clustering
```

```python
madrid_neighb.insert(0, 'Cluster Labels', kmeans.labels_)

madrid_merged = madrid_neighb[['Neighborhood', 'Neighborhood Latitude', 'Neighborhood Longitude', 'Cluster Labels', 'Num pools', 'Num gyms', 'Num ice cream shops', 'Num supermarkets', 'Num coworkings', 'Num monuments', 'Ratio']]
madrid_merged
```

```python
# create map
map_clusters = folium.Map(location=[latitude, longitude], zoom_start=11)

# set color scheme for the clusters
x = np.arange(kclusters)
ys = [i + x + (i*x)**2 for i in range(kclusters)]
colors_array = cm.rainbow(np.linspace(0, 1, len(ys)))
rainbow = [colors.rgb2hex(i) for i in colors_array]

# add markers to the map
markers_colors = []
for lat, lon, poi, cluster in zip(madrid_merged['Neighborhood Latitude'], madrid_merged['Neighborhood Longitude'], madrid_merged['Neighborhood'], madrid_merged['Cluster Labels']):
    label = folium.Popup(str(poi) + ' Cluster ' + str(cluster), parse_html=True)
    folium.CircleMarker(
        [lat, lon],
        radius=5,
        popup=label,
        color=rainbow[cluster-1],
        fill=True,
        fill_color=rainbow[cluster-1],
        fill_opacity=0.7).add_to(map_clusters)
       
map_clusters
```
